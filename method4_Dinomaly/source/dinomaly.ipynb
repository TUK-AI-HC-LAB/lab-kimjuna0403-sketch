```python
import os

# 파일 경로 자동으로 찾기 (Colab, Kaggle, 또는 현재 폴더 모두 커버)
paths_to_try = [
    'utils.py', 
    'Dinomaly/utils.py', 
    '/content/Dinomaly/utils.py', 
    '/kaggle/working/Dinomaly/utils.py'
]

target_file = None
for p in paths_to_try:
    if os.path.exists(p):
        target_file = p
        break

if target_file:
    with open(target_file, 'r') as f:
        content = f.read()
    
    # 낡은 append 코드를 최신 concat 문법으로 교체
    old_str = 'df = df.append({"pro": mean(pros), "fpr": fpr, "threshold": th}, ignore_index=True)'
    new_str = 'new_row = pd.DataFrame([{"pro": mean(pros), "fpr": fpr, "threshold": th}]); df = pd.concat([df, new_row], ignore_index=True)'
    
    if old_str in content:
        content = content.replace(old_str, new_str)
        with open(target_file, 'w') as f:
            f.write(content)
        print(f"✅ 성공: {target_file} 파일 수정 완료! 이제 마음 편히 모델을 돌리세요.")
    elif new_str in content:
        print(f"✅ 알림: 이미 {target_file} 파일이 수정되어 있습니다! 바로 돌리시면 됩니다.")
    else:
        print("⚠️ 주의: 해당 코드를 찾을 수 없습니다. (버전이 다르거나 이미 다른 방식으로 수정됨)")
else:
    print("❌ 에러: utils.py 파일을 아예 찾을 수 없습니다. git clone이 제대로 되었는지 확인해주세요.")
```
